```python
# Example 2.1 Descriptive Statistics

import scipy as sp
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets._samples_generator import make_blobs

X, y = make_blobs(n_samples = 300, centers = 4, cluster_std = 0.60, random_state = 0)
plt.scatter(X[:, 0], X[:, 1])

X_mean = np.mean(X[:,1])
print('Mean of X:', X_mean)

X_std = np.std(X[:,1])
print('Standard deviation of X:', X_std)

X_median = np.median(X[:,1])
print('Median of X:', X_median)

X_skewness = stats.skew(X[:,1])
print('Skewness of X:', X_skewness)

X_kurtosis = stats.kurtosis(X[:,1])
print('Kurtosis of X:', X_kurtosis)
```

```python
# Example 2.2: Min-Max Scaling Process

from sklearn import preprocessing
import numpy as np

X_train = np.array([[ 1., -1.,  2.],
                        [ 2.,  0.,  0.],
                        [ 0.,  1., -1.]])

print('Original Matrix:\n', X_train)
min_max_scaler = preprocessing.MinMaxScaler()
X_train_minmax = min_max_scaler.fit_transform(X_train)
print('Scaled Matrix:\n', X_train_minmax)
```

```python
# Example 2.3: Standardization Process

from sklearn import preprocessing
import numpy as np

X_train = np.array([[ 1., -1.,  2.],
                        [ 2.,  0.,  0.],
                        [ 0.,  1., -1.]])

X_scaled = preprocessing.scale(X_train)
print('Scaled Matrix:\n', X_scaled)

print(X_scaled.mean(axis=0))
print(X_scaled.std(axis=0))
```

```python
# Example 2.4: StandardScaler Process

from sklearn import preprocessing
import numpy as np

X_train = np.array([[ 2., -2,  5.],
                        [ 4.,  0.,  0.],
                        [ 0.,  2., -4.]])

print('Original Matrix:\n', X_train)
scaler = preprocessing.StandardScaler().fit(X_train)

scaler
print("Scalar Mean:", scaler.mean_)
print("Scalar:", scaler.scale_)
X_train_scaled = scaler.transform(X_train)
print('Scaled Matrix:\n', X_train_scaled)
```

```python
# Example 2.5: StandardScaler Process

from matplotlib import pyplot as plt
from sklearn.datasets._samples_generator import make_blobs
from sklearn import preprocessing
import numpy as np

X, y = make_blobs(n_samples = 100, centers = 4, cluster_std = 0.60, random_state = 0)
plt.scatter(X[:, 0], X[:, 1])

print("Original Matrix:\n", X)
scaler = preprocessing.StandardScaler().fit(X)
scaler

print("Scalar Mean:", scaler.mean_)
print("Scalar:", scaler.scale_)
X_scaled = scaler.transform(X)
print('Scaled Matrix:\n', X_scaled)
```
